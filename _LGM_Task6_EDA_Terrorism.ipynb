```python
countries_with_most_terrorism = df.country_txt.value_counts().head(10)
countries = list(countries_with_most_terrorism.index)
print(countries)
```

Visualizing the data

```python
fig, ax = plt.subplots(figsize=(12,5))
ax.bar(countries_with_most_terrorism.index,countries_with_most_terrorism.values, color='black')
plt.title('countries with most terrorism')
plt.xlabel('country')
plt.ylabel('values')
plt.show()
```

```python
year = df.iyear.value_counts().head(5)
plt.figure(figsize=(10,5))
plt.bar(year.index, year.values, color= 'purple')
plt.title('Years with most number of terrorist activity')
plt.xlabel('Year')
plt.ylabel('Values')
plt.show()
```

```python
target = df['targtype1_txt'].value_counts().head(10)
fig,ax = plt.subplots(figsize=(16,6))
ax.bar(target.index,target.values,color='g')
plt.title('Type of Targets')
plt.xlabel("Targets")
plt.ylabel("Values")
plt.show()
```

```python
weapon_type = df['weaptype1_txt'].value_counts().head(5)
fig, ax = plt.subplots(figsize=(12,8))
ax.bar(weapon_type.index, weapon_type.values)
plt.title('Types of attack')
plt.xlabel('Attacks')
plt.ylabel('values')
plt.show()
```

```python
sns.boxplot(x=df['iyear'])
```

```python
religious_target = df[df['targtype1_txt'] == 'Religious Figures/Institutions']
plt.figure(figsize=(10,6))
plt.bar(religious_target['country_txt'].value_counts().head().index,religious_target['country_txt'].value_counts().head().values,color='grey')
plt.title('countries with most terrorism on religious figures')
plt.xlabel('country')
plt.ylabel('values')
plt.show()
```

```python
cities_with_most_terrorism = df.city.value_counts().head(10)
cities = list(cities_with_most_terrorism.index)
print(cities)
plt.subplots(figsize=(10,6))
sns.countplot('city',data=df,order= df['city'].value_counts().index[1:10])
plt.xticks(rotation=90)
plt.xlabel("City")
plt.title(' Maximum Number Of Terrorist Activities By City')
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.countplot(data=df, x=df["region_txt"], palette='Set2')
plt.xticks(rotation=45)
plt.xlabel("Region",  size=13, fontweight="bold"  )
plt.ylabel("Count",  size=13, fontweight="bold" )
plt.title("count of attack on various regions in the world ", fontweight="bold" ,  size=25)
plt.show()
```

```python
sns.distplot(a=df['country'])
```

```python
sns.scatterplot(x=df['country'],y=df['iyear'])
```

```python
year = df['iyear'].unique()
years_count= df['iyear'].value_counts(dropna= False).sort_index()
plt.figure(figsize=(12,6))
sns.barplot(x=year, y=years_count, palette='colorblind' )
plt.xticks(rotation=60)
plt.xlabel('Attacking years',fontsize=18)
plt.ylabel('Number of attacks each year',fontsize=18)
plt.title('Attacks in year', fontsize=18)
plt.show()
```

```python
plt.subplots(figsize=(18,8))
sns.countplot(df["iyear"],order=df['iyear'].value_counts().index,palette="colorblind")
plt.xticks(rotation=90)
plt.xlabel("Attacktype",fontsize=15)
plt.ylabel("count",fontsize=15)
plt.title("Attack per year",fontsize=20)
plt.show()
```

```python
plt.figure(figsize = (11,7))
sns.barplot(y = df['gname'].value_counts()[1:11].values, x = df['gname'].value_counts()[1:11].index,palette='rocket')
plt.title('Top 10 Terrorist Organisation with Highest Terror Attacks',size=20, fontweight="bold")
plt.ylabel('Terror Group Name',size=15, fontweight="bold")
plt.xlabel('Number of Attacks',size=15, fontweight="bold")
plt.xticks(rotation= 70)
plt.show()
```

```python
df['country'].value_counts().to_frame().head(10).plot(kind='bar',color='g',figsize=(12,9))
plt.title('Top 10 terrorist groups',fontsize=18)
plt.xlabel('terrorist group count',fontsize=18)
plt.ylabel('Attack number',fontsize=18)
plt.show()
```

## Summary

```python
print('              SUMMARY                   ')
print('Country with most attacks: ',df['country_txt'].value_counts().idxmax())
print('Country with least attacks: ',df['country_txt'].value_counts().idxmin())
print('City with most attacks: ',df['city'].value_counts().index[1])
print("City with least attacks:",df['city'].value_counts().idxmin())
print("Region with the most attacks:",df['region_txt'].value_counts().idxmax())
print("Region with the least attacks:",df['region_txt'].value_counts().idxmin())
print("Year with the most attacks:",df['iyear'].value_counts().idxmax())
print("Year with the least attacks:",df['iyear'].value_counts().idxmin())
print("Month with the most attacks:",df['imonth'].value_counts().idxmax())
print("Month with the least attacks:",df['imonth'].value_counts().idxmin())
print("Most active terrorist group: ",df['gname'].value_counts()[1:11].idxmax() )
print("Most common Attack Types:",df['attacktype1_txt'].value_counts().idxmax())
```
